## 引入模組

```python
import pandas as pd #引入Pandas模組 as pd
import seaborn as sns
pd.set_option("display.max_rows", 1000)    #設定最大能顯示1000rows
pd.set_option("display.max_columns", 1000) #設定最大能顯示1000columns

from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  
# 指定默認字形：解決plot不能顯示中文問題
mpl.rcParams['axes.unicode_minus'] = False
```

## 讀取資料

```python
df = pd.read_excel('1.xlsx')
```

## 資料類別

```python
df.dtypes
```

## 設定Index

```python
df1 = df.set_index(['觀測時間'])
```

## 顯示所有物種總數量

```python
dfTotal=df1.groupby('物種中文名').sum()
```

```python
dfTotal
```

## 共8607隻鳥

```python
dfTotal.sum()
```

## 春天的鳥類

```python
dfSpring=df1.loc['春天']
```

```python
dfSpringSort=dfSpring.sort_values(by='物種中文名').head(1000)
```

```python
dfSpringSort.groupby('物種中文名').sum()
```

## 夏天的鳥類

```python
dfSummer=df1.loc['夏天']
```

```python
dfSummerSort=dfSummer.sort_values(by='物種中文名').head(1000)
```

```python
dfSummerSort.groupby('物種中文名').sum()
```

## 秋天的鳥類

```python
dfFall=df1.loc['秋天']
```

```python
dfFallSort=dfFall.sort_values(by='物種中文名').head(1000)
```

```python
dfFallSort.groupby('物種中文名').sum()
```

## 冬天的鳥類

```python
dfWinter=df1.loc['冬天']
```

```python
dfWinterSort=dfWinter.sort_values(by='物種中文名').head(1000)
```

```python
dfWinterSort.groupby('物種中文名').sum()
```

```python
dfSpringSort.groupby('物種中文名').sum().to_excel('春天的鳥類.xlsx', encoding='utf8')
dfSummerSort.groupby('物種中文名').sum().to_excel('夏天的鳥類.xlsx', encoding='utf8')
dfFallSort.groupby('物種中文名').sum().to_excel('秋天的鳥類.xlsx', encoding='utf8')
dfWinterSort.groupby('物種中文名').sum().to_excel('冬天的鳥類.xlsx', encoding='utf8')
```
